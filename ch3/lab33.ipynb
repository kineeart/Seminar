```python
#Import pandas, numpy and matplotlib.pyplot
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# Create a pandas DataFrame with 12 months of imaginary sales data for a tech store
data = {
    "Month": [
        "January", "February", "March", "April", "May", "June",
        "July", "August", "September", "October", "November", "December"
    ],
    "Sales": [
        15000, 18000, 22000, 25000, 28000, 30000,
        32000, 31000, 29000, 26000, 23000, 20000
    ]
}
df = pd.DataFrame(data)

```

```python
# Plot a bar chart showing the sales data. Add title and labels
plt.figure(figsize=(10, 6))
plt.bar(df['Month'], df['Sales'])
plt.title('Monthly Sales Data')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
